```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('../data/hypothesis-test-data.csv')
df
```

```python
df = df[df['Male_Deaths'] >= 5]
df = df[df['Female_Deaths'] >= 5]
```

```python
len(df)
```

```python
df['Male_death_ratio'] = df['Male_Deaths']/df['Male_cases']
df['Female_death_ratio'] = df['Female_Deaths']/df['Female_cases']
```

```python
df['P'] = (df['Female_death_ratio']*df['Female_cases'] + df['Male_death_ratio']*df['Male_cases'])/(df['Female_cases'] + df['Male_cases'])
df['SE'] = (df['P']*(1-df['P'])*(1/df['Female_cases'] + 1/df['Male_cases']))**0.5
df['z'] = (df['Female_death_ratio'] - df['Male_death_ratio'])/df['SE']
df['p_value'] = stats.norm.sf(abs(df['z']))*2
```

```python
df['difference_ratio'] = df['Female_death_ratio'] - df['Male_death_ratio']
df['95_CI_low'] = df['difference_ratio'] - df['SE']
df['95_CI_high'] = df['difference_ratio'] + df['SE']
```

```python
df[df['p_value'] <= 0.05]
```

```python
df[df['p_value'] <= 0.1]
```
